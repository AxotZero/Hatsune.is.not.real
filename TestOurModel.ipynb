```python
import tensorflow as tf
new_model = tf.keras.models.load_model('fake_real_model.model')
```

```python
##test result
import matplotlib.pyplot as plt
import numpy as np
import cv2

Dir = 'D:\Course_2-2\python\isThatReal\\testData'
Class = ['\\fake', '\\real']

for i in Class:
    for j in range(5):
        targetDir = Dir + i + str(j) + '.jpg'
        img = cv2.imread(targetDir)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img, cmap='hsv')
        plt.show()
        
        imgArray = cv2.imread(targetDir, cv2.IMREAD_GRAYSCALE)
        imgArray = cv2.resize(imgArray,(70,70))
        imgArray = imgArray / 255.0
        #plt.imshow(imgArray, cmap=plt.cm.binary)
        #plt.show()
        imgArray = imgArray.reshape(-1, 70, 70, 1)
        prediction = new_model.predict(imgArray)
        if(prediction[0][0] < 0.5):
            print('It\'s fake with %f percent' %((1-prediction[0][0]) * 100))
            print('')
        else:
            print('%d%%' %(prediction[0][0] * 100))
## < 0.5 is fake, > 0.5 is real
```
